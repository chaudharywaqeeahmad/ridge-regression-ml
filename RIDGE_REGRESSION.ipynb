```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import pandas as pd # Import pandas, commonly used for data
import numpy as np # Import numpy for creating example data

# --- FIX START ---
# The error "NameError: name 'X' is not defined" indicates that the variables
# X (features) and y (target) have not been defined before being used in
# train_test_split. You need to load or create your dataset first.

# As an example, let's create some dummy data for X and y:
# In a real scenario, you would load your data, for instance:
# df = pd.read_csv('your_data.csv')
# X = df[['feature_col1', 'feature_col2']] # Your feature columns
# y = df['target_col'] # Your target column

# Dummy data for demonstration:
np.random.seed(42) # for reproducibility
X = pd.DataFrame(np.random.rand(100, 5), columns=[f'feature_{i}' for i in range(5)])
y = pd.Series(np.random.rand(100) * 10 + 2 * X['feature_0'] + np.random.randn(100) * 2)
# --- FIX END ---

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Create a Ridge regression object
ridge = Ridge(alpha=1.0)  # Regularization parameter alpha is 1.0

# Fit the Ridge regression model
ridge.fit(X_train, y_train)

# Predict the target variable for the test data
y_pred = ridge.predict(X_test)

# Calculate the root mean squared error (RMSE)
# The 'squared=False' argument was introduced in scikit-learn version 0.22
# For compatibility, we calculate MSE and then take its square root.
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Output the RMSE
print("Root Mean Squared Error:", rmse)
```
